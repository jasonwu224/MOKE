```python
df_calibration = pd.read_csv(r"C:\Users\gxlab\Code\MOKE\data\calibration\Volts_to_Gauss_Calibration.csv")
df_calibration.head()
```

```python
from scipy.stats import linregress
volt = df_calibration["Voltage (V)"]
mag = df_calibration["Magnetic Flux Density (G)"]
slope, intercept, r_value, p_value, std_err = linregress(volt, mag)
print(slope, intercept, r_value, p_value, std_err)
fitted_mag = slope*volt+intercept
plt.scatter(volt, mag)
plt.plot(volt, fitted_mag, color="red", label=f"y = {slope:.2f}x{intercept:.2f}, r^2 = {r_value**2:.5f}")
plt.legend()
plt.xlabel("Voltage (V)")
plt.ylabel("Magnetic Flux Density (G)")
```

```python
# converts volts to gauss using fitted regression line from calibration data
# does not modify original data
def volts_to_gauss(volts):
    slope = 1.85
    intercept = -1.20
    gauss = slope*volts.copy() + intercept
    return gauss
```

```python
df_dict = {}
paths = (r"C:\Users\gxlab\Code\MOKE\data\2025-07-17_14-56-28\H+_data.csv", r"C:\Users\gxlab\Code\MOKE\data\2025-07-21_11-09-05\H+_data.csv", r"C:\Users\gxlab\Code\MOKE\data\2025-07-21_12-34-02\H+_data.csv", r"C:\Users\gxlab\Code\MOKE\data\2025-07-21_12-55-48\data_2025-07-21_12-55-48.csv")
for i, path in enumerate(paths):
    df_dict[f"{i+1} deg off extinction"] = pd.read_csv(path, comment="#")

df_dict
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 7))
ax = ax.flatten()
for i, (key, df) in enumerate(df_dict.items()):
    ax[i].plot(volts_to_gauss(df["Voltage (V)"]), df["Average Intensity"])
    ax[i].set_title(key)
    ax[i].set_xlabel("Magnetic field (G)")
    ax[i].set_ylabel("Avg Intensity (Counts)")
fig.suptitle("Hysteresis Loops for Incident P-Polarized Light")
fig.tight_layout()
```

```python
df = df_dict["4 deg off extinction"]
def analysis(df):
    n = len(df["Voltage (V)"])
    x = df["Voltage (V)"]
    I0 = df["Average Intensity"][0]
    y = df["Average Intensity"]
    plt.plot(x, y)
    pos_sweep_avg_I = y[0:2*n//5].mean()
    neg_sweep_avg_I = y[2*n//5:4*n//5].mean()
    total_avg = y[0:4*n//5].mean()
    volts = np.arange(-35, 35, 0.5)
    plt.plot(volts, np.full_like(volts, pos_sweep_avg_I))
    plt.plot(volts, np.full_like(volts, neg_sweep_avg_I))
    plt.plot(volts, np.full_like(volts, I0))
    plt.plot(volts, np.full_like(volts, total_avg))
    percent_change_I = 100*(neg_sweep_avg_I - pos_sweep_avg_I)/I0
    print(f"I_0: {I0:.3f}")
    print(f"Low I: {pos_sweep_avg_I:.3f}")
    print(f"High I: {neg_sweep_avg_I:.3f}")
    print(f"Average I: {total_avg:.3f}")
    print(f"Discrepency between I_0 and Average I: {100*(total_avg-I0)/I0:.3f} %")
    print(f"Percent change in intensity: {percent_change_I:.3f} %")
analysis(df)
```
