```python
df_calibration = pd.read_csv(r"C:\Users\gxlab\Code\MOKE\data\calibration\Volts_to_Gauss_Calibration.csv")
df_calibration.head()
```

```python
from scipy.stats import linregress
volt = df_calibration["Voltage (V)"]
mag = df_calibration["Magnetic Flux Density (G)"]
slope, intercept, r_value, p_value, std_err = linregress(volt, mag)
print(slope, intercept, r_value, p_value, std_err)
fitted_mag = slope*volt+intercept
plt.scatter(volt, mag)
plt.plot(volt, fitted_mag, color="red", label=f"y = {slope:.2f}x{intercept:.2f}, r^2 = {r_value**2:.5f}")
plt.legend()
plt.xlabel("Voltage (V)")
plt.ylabel("Magnetic Flux Density (G)")
```

```python
# converts volts to gauss using fitted regression line from calibration data
# does not modify original data
def volts_to_gauss(volts):
    slope = 1.85
    intercept = -1.20
    gauss = slope*volts.copy() + intercept
    return gauss
```

```python
df = df_dict["4 deg off extinction"]
def analysis(df):
    n = len(df["Voltage (V)"])
    x = df["Voltage (V)"]
    I0 = df["Average Intensity"][0]
    y = df["Average Intensity"]
    plt.plot(x, y)
    pos_sweep_avg_I = y[0:2*n//5].mean()
    neg_sweep_avg_I = y[2*n//5:4*n//5].mean()
    total_avg = y[0:4*n//5].mean()
    volts = np.arange(-35, 35, 0.5)
    plt.plot(volts, np.full_like(volts, pos_sweep_avg_I))
    plt.plot(volts, np.full_like(volts, neg_sweep_avg_I))
    plt.plot(volts, np.full_like(volts, I0))
    plt.plot(volts, np.full_like(volts, total_avg))
    percent_change_I = 100*(neg_sweep_avg_I - pos_sweep_avg_I)/I0
    print(f"I_0: {I0:.3f}")
    print(f"Low I: {pos_sweep_avg_I:.3f}")
    print(f"High I: {neg_sweep_avg_I:.3f}")
    print(f"Average I: {total_avg:.3f}")
    print(f"Discrepency between I_0 and Average I: {100*(total_avg-I0)/I0:.3f} %")
    print(f"Percent change in intensity: {percent_change_I:.3f} %")
analysis(df)
```

```python
from matplotlib.collections import LineCollection
from matplotlib.colors import Normalize
import matplotlib.cm as cm
# Helper function: create color-gradient line segments
def colored_line(ax, x, y, cmap='viridis'):
    points = np.array([x, y]).T.reshape(-1, 1, 2)
    segments = np.concatenate([points[:-1], points[1:]], axis=1)
    t = np.linspace(0, 1, len(x))  # time-normalized
    norm = Normalize(0, 1)
    lc = LineCollection(segments, cmap=cmap, norm=norm)
    lc.set_array(t)
    lc.set_linewidth(2)
    line = ax.add_collection(lc)
    return line
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection
from matplotlib.colors import LinearSegmentedColormap, Normalize
import matplotlib.cm as cm

# ["#5ebcff", "#082a42"] blue gradient
color_map = LinearSegmentedColormap.from_list("custom_gray_gold", ["#40b3ff", "#082a42"])
# Helper function: draw a color-gradient line using LineCollection
def colored_line(ax, x, y, cmap=color_map):
    points = np.array([x, y]).T.reshape(-1, 1, 2)
    segments = np.concatenate([points[:-1], points[1:]], axis=1)
    t = np.linspace(0, 1, len(x))  # normalized "time"
    norm = Normalize(0, 1)
    lc = LineCollection(segments, cmap=cmap, norm=norm)
    lc.set_array(t)
    lc.set_linewidth(2)
    ax.add_collection(lc)
    return lc

# Volts to gauss conversion
def volts_to_gauss(volts):
    return 1.85 * volts - 1.20

# Load data
df_dict = {}
names = ("Dither only", "Diffuse Only", "Dither & Diffuse")
paths = (
    r"data\2025-07-28_12-03-32_no_diffuse\data_2025-07-28_12-03-32.csv",
    r"data\2025-07-28_11-56-04_no_dither\data_2025-07-28_11-56-04.csv",
    r"data\2025-07-28_11-39-29_dither_and_diffuse\data_2025-07-28_11-39-29.csv"
)

for name, path in zip(names, paths):
    df_dict[name] = pd.read_csv(path, comment="#")

# Setup plots
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax = ax.flatten()

cbar_ref = None  # To link a colorbar to one line collection

for i, (key, df) in enumerate(df_dict.items()):
    n = len(df)
    B = volts_to_gauss(df["Voltage (V)"])
    I = df["Average Intensity"]

    lc = colored_line(ax[i], B, I)
    if cbar_ref is None:
        cbar_ref = lc  # Only need one reference for shared colorbar

    ax[i].set_title(key)
    ax[i].set_xlabel("Magnetic field (G)")
    ax[i].set_ylabel("Avg Intensity (Counts)")
    ax[i].autoscale() 

fig.suptitle("Hysteresis Loops for Incident P-Polarized Light")
fig.text(
    0.5, -0.2,
    "Comparison of despeckling methods. 2 Â± 0.5Â° off extinction, 4 ms exposure, 15-frame average.\nDithering: 240 Hz at 250 mVpp.",
    ha="center"
)
fig.tight_layout()
# Place a horizontal colorbar below the subplots
cbar_ax = fig.add_axes([0.25, 0, 0.5, 0.02])  # [left, bottom, width, height]
fig.colorbar(cbar_ref, cax=cbar_ax, orientation='horizontal', label="Relative Time in Sweep")
```
